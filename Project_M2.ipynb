```python
# Change column name "Content Rating" --> "Age Rating"
PL_data = PL_data.rename(columns={'Content Rating': 'Age Rating'})
```

```python
# Change column name "Type" --> "pricing_type"
PL_data = PL_data.rename(columns={'Type': 'Pricing_type'})

```

```python
#standardize Pricing_type to only FREE and PAID
PL_data['Pricing_type'] = PL_data['Pricing_type'].replace({
    'free': 'FREE',
    '0': 'FREE',
    'paid': 'PAID',
    'nan': np.nanπ
})
```

```python
#standardize Age Rating
rating_map = {
    'Everyone': 'All',
    'Everyone 10+': '10+',
    'Teen': '13+',
    'Mature 17+': '17+',
    'Adults only 18+': '18+',
    'Unrated': 'NO Rating',
    'nan': np.nan
}

PL_data['Age Rating'] = PL_data['Age Rating'].map(rating_map)
```

```python
# removing unwanted category 
PL_data = PL_data[PL_data['Category'] != '1.9']
```

```python
# convert reviews from string to int
PL_data['Reviews'] = pd.to_numeric(PL_data['Reviews'], errors='coerce').astype('Int64')
```

```python
#standardize 'Size' column
#Converted 'Size' to kilobytes, handling different units.

def convert_size(size):
    if pd.isna(size):
        return np.nan
    size = str(size).strip()
    
    if size == 'Varies with device':
        return np.nan
    elif size.endswith('M'):
        return float(size.replace('M', '')) * 1024
    elif size.endswith('k'):
        return float(size.replace('k', ''))
    else:
        return np.nan

PL_data['Size_KB'] = (
    PL_data['Size']
    .apply(convert_size)
    .round()
    .astype('Int64')
)


#
PL_data['Size_KB'] = PL_data['Size_KB'].astype('Int64')
PL_data['Size_KB'].dtype
```

```python
#converting "Price" column from string to int
PL_data['Price'] = (
    PL_data['Price']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.strip()
)
PL_data['Price'] = pd.to_numeric(PL_data['Price'], errors='coerce')

```

```python
#updating "Price" 
PL_data.loc[PL_data['Price'] == 0, 'Pricing_type'] = 'Free'
```

```python
#renaming "Price"
PL_data = PL_data.rename(columns={'Price': 'Price in dollar'})
```

```python
#Converted 'Last Updated' to datetime format.
PL_data['Last Updated'] = pd.to_datetime(PL_data['Last Updated'], errors='coerce')
PL_data.loc[:,'Last Updated']
```

```python
#Parsed and converted 'Installs' and 'Price' fields to numeric values.
installs_str = PL_data['Installs'].astype(str).str.replace(',', '', regex=False).str.replace('+', '', regex=False)
PL_data['Installs_num'] = pd.to_numeric(installs_str, errors='coerce')
```

```python
#updating columns name
PL_data.columns = (
    PL_data.columns
    .str.strip()
    .str.title()
    .str.replace(' ', '_')
)

```

```python
# dropping "Size" column
PL_data = PL_data.drop(columns=['Size'])
```

```python
# filling null values of 'Current Ver' column
PL_data.loc[:,'Current_Ver'] = PL_data.loc[:,'Current_Ver'].fillna( 'Varies with device')

```

```python
# filling null values of 'Android Ver' column
PL_data.loc[:,'Android_Ver'] = PL_data.loc[:,'Android_Ver'].fillna( 'Varies with device')

```

```python
# Analyze relationship between price and installs using cleaned columns already created (eda_df)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

rel_df = PL_data.copy()
rel_df = rel_df[(rel_df['Installs_Num'].notna()) & (rel_df['Price_In_Dollar'].notna())]
rel_df['is_paid'] = rel_df['Price_In_Dollar'] > 0

# Summary: free vs paid installs (use log because installs are heavy-tailed)
summary_fp = rel_df.groupby('is_paid').agg(
    apps=('App','count'),
    median_installs=('Installs_Num','median'),
    mean_installs=('Installs_Num','mean'),
    median_log_installs=('Installs_Num', lambda s: np.median(np.log10(s + 1))),
    mean_log_installs=('Installs_Num', lambda s: np.mean(np.log10(s + 1)))
).reset_index()
print(summary_fp)

# Price bins for paid apps only
paid_df = rel_df[rel_df['is_paid']].copy()
price_bins = [-0.01, 0, 0.99, 2.99, 4.99, 9.99, 19.99, 49.99, 1000]
price_labels = ['Free', '0.01-0.99', '1-2.99', '3-4.99', '5-9.99', '10-19.99', '20-49.99', '50+']
rel_df['price_bin'] = pd.cut(rel_df['Price_In_Dollar'], bins=price_bins, labels=price_labels)

bin_summary = rel_df.groupby('price_bin', dropna=False).agg(
    apps=('App','count'),
    median_installs=('Installs_Num','median'),
    median_log_installs=('Installs_Num', lambda s: np.median(np.log10(s + 1)))
).reset_index()
print(bin_summary)

# Correlation among paid apps (use log installs)
paid_df = paid_df[(paid_df['Price_In_Dollar'] > 0) & (paid_df['Installs_Num'] >= 0)]

paid_df['log_installs'] = np.log10(paid_df['Installs_Num'] + 1)
paid_df['log_price'] = np.log10(paid_df['Price_In_Dollar'] + 1e-9)
pearson_corr = paid_df[['Price_In_Dollar','log_installs']].corr(method='pearson').iloc[0,1]
spearman_corr = paid_df[['Price_In_Dollar','log_installs']].corr(method='spearman').iloc[0,1]
print(pearson_corr)
print(spearman_corr)

# Visual 1: boxenplot of log installs by price bin
plot_df = rel_df[rel_df['price_bin'].notna()].copy()
plot_df['log_installs'] = np.log10(plot_df['Installs_Num'] + 1)

plt.figure(figsize=(11,4.5))
sns.boxenplot(data=plot_df, x='price_bin', y='log_installs', color='cornflowerblue')
plt.xticks(rotation=25, ha='right')
plt.xlabel('Price bin (USD)')
plt.ylabel('log10(installs + 1)')
plt.title('Installs vs Price (log scale)')
plt.tight_layout()
plt.show()

# Visual 2: scatter for paid apps (price vs log installs) with trendline
scatter_df = paid_df[(paid_df['Price_In_Dollar'] <= 100) & (paid_df['Installs_Num'] > 0)].copy()
plt.figure(figsize=(7.5,4.5))
sns.regplot(data=scatter_df, x='Price_In_Dollar', y='log_installs', scatter_kws={'alpha':0.25, 's':20}, line_kws={'color':'black'})
plt.xlabel('Price ($)')
plt.ylabel('log10(installs + 1)')
plt.title('Paid apps: Price vs Installs (log scale)')
plt.tight_layout()
plt.show()
```

## Final conclusion (Price vs Installs)
The strongest relationship is simply whether an app is free or paid. In this Play Store dataset, paid apps get dramatically fewer installs than free apps.

Free apps have a median of ~100,000 installs, while paid apps have a median of ~1,000 installs. That’s roughly a 100x drop at the median just from charging anything at all.

```python
# - Defined candidate file paths for Play Store data.
# - Checked for existing files and printed the list.
# - Loaded user reviews from a specified CSV file and displayed the first few rows.
# - Attempted to load the Play Store dataset from existing files, checking for the presence of an 'App' column and sufficient columns.
# - Recorded load attempts and their outcomes in a DataFrame.
# - Printed the results of the loading attempts and the loaded dataset if successful.
import os
import pandas as pd

candidate_paths = ['Play Store Data.csv', 'play_store_data.csv', 'PlayStoreData.csv', 'paste.txt', 'paste1.txt']
existing_paths = [p for p in candidate_paths if os.path.exists(p)]
print(pd.Series(existing_paths))

# Load user reviews (expected CSV)
user_reviews_df = pd.read_csv('User Reviews.csv', encoding='utf-8')
print(user_reviews_df.head())

# Try to load play store data from whichever file parses as a wide CSV with an App column
play_store_df = None
load_attempts = []

for path in existing_paths:
    try:
        try_df = pd.read_csv(path, encoding='utf-8', sep=None, engine='python')
        load_attempts.append([path, try_df.shape[0], try_df.shape[1], ','.join([c for c in try_df.columns[:5].astype(str)])])
        if 'App' in try_df.columns and try_df.shape[1] >= 8:
            play_store_df = try_df
            play_store_source = path
            break
    except Exception as exc:
        load_attempts.append([path, None, None, str(exc)[:120]])

attempts_df = pd.DataFrame(load_attempts, columns=['file','rows','cols','note'])
print(attempts_df)

if play_store_df is None:
    print('No Play Store dataset found in the provided files.')
else:
    print('Loaded Play Store dataset from: ' + play_store_source)
    print(play_store_df.head())
```

```python
# - Copied user reviews DataFrame
# - Converted sentiment polarity to numeric and dropped NaN values
# - Mapped sentiment labels to proxy ratings
# - Binned sentiment polarity into intervals
# - Summarized counts and means for each bin
# - Converted bins to strings and sorted them
# - Plotted average proxy ratings against sentiment polarity bins
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')

df_cmp2 = user_reviews_df.copy()
df_cmp2['Sentiment_Polarity'] = pd.to_numeric(df_cmp2['Sentiment_Polarity'], errors='coerce')
df_cmp2 = df_cmp2.dropna(subset=['Sentiment_Polarity'])

sent_to_star2 = {'Negative': 1, 'Neutral': 3, 'Positive': 5}
df_cmp2['rating_proxy'] = df_cmp2['Sentiment'].map(sent_to_star2)
df_cmp2 = df_cmp2.dropna(subset=['rating_proxy'])

# Bin polarity and prepare summary
bin_edges = np.linspace(-1, 1, 11)
df_cmp2['polarity_bin'] = pd.cut(df_cmp2['Sentiment_Polarity'], bins=bin_edges, include_lowest=True)

bin_summary2 = df_cmp2.groupby('polarity_bin', observed=True)['rating_proxy'].agg(['count','mean']).reset_index()
bin_summary2['bin_left'] = bin_summary2['polarity_bin'].apply(lambda x: x.left)
bin_summary2 = bin_summary2.sort_values('bin_left')
bin_summary2['polarity_bin_str'] = bin_summary2['polarity_bin'].astype(str)

print(bin_summary2[['polarity_bin_str','count','mean']].head(12))

plt.figure(figsize=(10,4))
sns.lineplot(data=bin_summary2, x='polarity_bin_str', y='mean', marker='o')
plt.xticks(rotation=45, ha='right')
plt.ylim(1,5)
plt.title('Average proxy rating vs sentiment polarity (binned)')
plt.ylabel('Mean proxy rating (1-5 from Sentiment label)')
plt.xlabel('Sentiment polarity bin')
plt.show()
```
